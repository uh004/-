### 최근 10,000개 데이터를 수집하여 seoul_real_estate.csv에 저장하기

```python
# 필요한 라이브러리
import requests
import json
import pandas as pd

# 인증키 URL
API_KEY = '6a4e4a656d716f74363061586f7967'

def main():
  data = pd.DataFrame()
  for i in range(1, 30):
    # tbLnOpendataRtmsV 데이터세트의 최근 1000*11 데이터 요청
    # 1 -> 1, 1000 / 2 -> 1001, 2000 / 3 -> 2001, 3000
    url = 'http://openapi.seoul.go.kr:8088/{}/json/tbLnOpendataRtmsV/{}/{}/'.format(API_KEY, 1 + ((i - 1)) * 1000 , i * 1000)
    req = requests.get(url)
    content = req.json()
    con = content['tbLnOpendataRtmsV']['row']
    result = pd.DataFrame(con)
    data = pd.concat([data, result])

  data = data.reset_index(drop=True)
  data['DEAL_YMD'] = pd.to_datetime(data['DEAL_YMD'], format=('%Y%m%d'))
  data.to_csv('seoul3_real_estate_1000.csv', index=False)

if __name__ == '__main__':
  main()
```
